## **Task 2 — NumPy: Gene Expression Analysis (TCGA)**

Gene expression data from RNA sequencing (RNA-seq) is typically represented as a matrix where **rows correspond to genes** and **columns to samples**. Each value represents the normalised expression level of a gene in a sample (here in $log_2$ TPM units).

The [TCGA](https://www.cancer.gov/ccg/research/genome-sequencing/tcga) (The Cancer Genome Atlas) is one of the largest cancer genomics projects, providing matched tumour and normal tissue samples for many cancer types. Identifying genes that are differentially expressed between tumour and normal tissue is a fundamental step in cancer biology.

You are provided with two files:
- `data/expression_matrix.csv`: rows = genes (Ensembl IDs), columns = samples. First column is `gene_id`.
- `data/sample_metadata.csv`: maps each sample ID to its condition (`tumor` or `normal`).

The following code reads both files:

````python
import numpy as np
import matplotlib.pyplot as plt

# Read expression matrix
import csv

with open('data/expression_matrix.csv', newline='') as f:
    reader = csv.reader(f)
    header = next(reader)
    sample_ids = header[1:]            # list of sample IDs
    gene_ids   = []
    expr_rows  = []
    for row in reader:
        gene_ids.append(row[0])
        expr_rows.append([float(x) for x in row[1:]])

expr_matrix = np.array(expr_rows)     # shape: (n_genes, n_samples)
gene_ids    = np.array(gene_ids)
sample_ids  = np.array(sample_ids)

# Read sample metadata
import pandas as pd
metadata = pd.read_csv('data/sample_metadata.csv')

# Boolean masks for tumor and normal samples
tumor_mask  = np.array([s in metadata[metadata['condition']=='tumor']['sample_id'].values
                         for s in sample_ids])
normal_mask = np.array([s in metadata[metadata['condition']=='normal']['sample_id'].values
                         for s in sample_ids])

print(f"Expression matrix shape: {expr_matrix.shape}")
print(f"Tumor samples: {tumor_mask.sum()} | Normal samples: {normal_mask.sum()}")
````
