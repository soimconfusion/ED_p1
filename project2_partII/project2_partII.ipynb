Genome-wide association studies (GWAS) systematically scan the human genome to find common genetic variants (SNPs) associated with complex traits and diseases in large population cohorts. 
The [GWAS Catalogue](https://www.ebi.ac.uk/gwas/) is the most comprehensive public repository of published GWAS results.

The file `data/gwas_catalog.tsv` contains a curated subset of the GWAS Catalogue. 
Below you can find a description of the columns in the file.

| Column | Description |
|--------|-------------|
| `STUDY ACCESSION` | Unique study identifier (e.g. `GCST0001234`) |
| `DISEASE/TRAIT` | Trait or disease under study |
| `CHR_ID` | Chromosome of the associated variant (1–22) |

| `P-VALUE` | Association p-value (string in scientific notation) |
| `OR or BETA` | Effect size (odds ratio or beta coefficient) |
| `MAPPED_GENE` | Gene(s) mapped to the variant |
| `PUBMEDID` | PubMed ID of the source publication |
