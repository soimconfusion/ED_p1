## **Parte I — DNA Sequence Processing**

### **Task 1 (T1):** Read multiple FASTA sequences

The FASTA format is the universal standard for representing biological sequences. Each entry begins with a header line (`>`) followed by one or more lines containing the sequence.

Complete the function `readFASTA` which reads a file in FASTA format containing multiple sequences.

A simplified generic example of a FASTA file is shown below:

```python
>seq1 Homo sapiens gene fragment
ATGCGTAGCTAGCTAG
>seq2 Mus musculus gene fragment
ATGCGTGGGCTAG
```

The expected result of the `readFASTA` function should be a dictionary `{ ID : (desc, seq) }`, where the key `ID` represents the sequence identifier (the first word of the header line, i.e., immediately after `>`) and the value is a tuple `(desc, seq)`.

Here, `desc` corresponds to the remaining text of the header line (which may be empty), and `seq` is the concatenation of all subsequent sequence lines with spaces and newline characters removed. (*Note* that each sequence may span multiple lines.)

For the example above, the function should return the following dictionary:

```python
{
 "seq1": ('Homo sapiens gene fragment', 'ATGCGTAGCTAGCTAG'),
 "seq2": ('Mus musculus gene fragment', 'ATGCGTGGGCTAG')
}
```

### **Task 5 (T5):** Reverse complement

In DNA molecules, each strand has a complementary strand formed by specific base pairing. The bases pair as follows

| base  | complement |
| ------------- | ------------- |
| `A`  | `T`  |
| `T`  | `A`  |
| `C`  | `G`  |
| `G`  | `C`  |

In addition, the two DNA strands are antiparallel, meaning that the complementary sequence must be read in the reverse direction.

For example, consider the DNA sequence `ATGCCGTA`. Its base-by-base complement is `TACGGCAT`. The **reverse complement** corresponds to the complementary sequence read in reverse order, which in this case is also `TACGGCAT`. (*Note:* n this particular example it coincides with the direct complement, but this will not usually be the case.)

Complete the function `reverseComplement` which receives a DNA sequence and returns its reverse complement.

For example:
```python
dna_seq = Seq("ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCGATAG")

# Complemento
rev_complement = reverseComplement(dna_seq)
print("Reverse complement:", rev_complement)

CTATCGCACCCTTTCAGCGGCCCATTACAATGGCCAT
```
